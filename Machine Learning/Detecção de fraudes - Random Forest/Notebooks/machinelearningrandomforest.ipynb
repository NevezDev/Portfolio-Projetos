#Importando as bibliotecas necess√°rias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_curve, auc, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

# Carregando o conjunto de dados e exibindo

```python
data = pd.read_csv('creditcard.csv')
data
```

# Iniciando o pr√©-processamento dos dados.

```python
#Exibe informa√ß√µes sobre o conjunto de dados, como tipos de colunas e valores n√£o nulos.
data.info()
```

```python
#Mostrando estat√≠sticas descritivas
data.describe().T
```

```python
#Contando a frequ√™ncia de cada valor na coluna 'Class', que provavelmente representa se uma transa√ß√£o √© fraudulenta ou n√£o.
data['Class'].value_counts()
```

```python
# Verificando a quantidade de valores ausentes em cada coluna do conjunto de dados.
data.isnull().sum()
```

Dividindo os dados em features (x) e target (y), onde x cont√©m todas as colunas exceto a √∫ltima e y cont√©m apenas a √∫ltima coluna ('Class').

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
y
```

Dividindo os dados em conjuntos de treinamento e teste usando train_test_split, com 20% dos dados para teste e estratifica√ß√£o para garantir a mesma propor√ß√£o de classes em ambos os conjuntos.

```python
x_train, x_test, y_train, y_test = train_test_split( x , y, test_size= 0.2, stratify= y, random_state= 42)
```

```python
# Verificando se h√° valores ausentes no conjunto de treinamento de (x_train).
x_train.isnull().sum()
```

```python
# Fazendo o mesmo com o conjunto alvo
y_train.isnull().sum()
```

```python
# Contando a frequ√™ncia de cada valor no conjunto de treino (y_train).
y_train.value_counts()
```

```python
# Fazendo o mesmo com o conjunto alvo de testes
y_test.value_counts()
```

Criando um modelo RandomForestClassifier com pesos balanceados para as classes e o treinando com os dados de treinamento.

```python
modelo = RandomForestClassifier(class_weight='balanced', random_state=89)
modelo.fit(x_train, y_train)
```

Fazendo previs√µes no conjunto de teste usando o modelo treinado

```python
previsao = modelo.predict(x_test)
previsao
```

# Avaliando o desempenho do modelo.

```python
# Calculando a precis√£o, recall e AUC
precisao, recall, _ = precision_recall_curve(y_test, previsao)
auc = auc(recall, precisao)
auc
```

```python
 # Imprimindo o relat√≥rio de classifica√ß√£o, que inclui precis√£o, recall, F1-score e suporte para cada classe.
classificacao = classification_report(y_test, previsao)
print(classificacao)
```

 Calculando a precis√£o, recall e AUC e plotando a curva de precis√£o-recall para visualizar o desempenho do modelo.

```python
from sklearn.metrics import precision_recall_curve, auc # Make sure to import the auc function

precisao, recall, _ = precision_recall_curve(y_test, previsao)
auc_score = auc(recall, precisao) # Use the imported auc function instead of the overwritten variable

plt.plot(recall, precisao, label='AUC = {:.2f}'.format(auc_score))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curva de Precis√£o-Recall')
plt.legend(loc='lower left')
plt.show()
```

 Calculando a acur√°cia do modelo no conjunto de teste.

```python
modelo.score(x_test, y_test)
```

Criando e exibindo a matriz de confus√£o para visualizar o desempenho do modelo em termos de verdadeiros positivos, falsos positivos, verdadeiros negativos e falsos negativos.

```python
cm = confusion_matrix(y_test, previsao)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Previs√£o')
plt.ylabel('Valor Real')
plt.title('Matriz de Confus√£o')
plt.show()
```
