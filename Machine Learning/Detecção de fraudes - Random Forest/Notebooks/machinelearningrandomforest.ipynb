#Importando as bibliotecas necessárias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_curve, auc, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

# Carregando o conjunto de dados e exibindo

```python
data = pd.read_csv('creditcard.csv')
data
```

# Iniciando o pré-processamento dos dados.

```python
#Exibe informações sobre o conjunto de dados, como tipos de colunas e valores não nulos.
data.info()
```

```python
#Mostrando estatísticas descritivas
data.describe().T
```

```python
#Contando a frequência de cada valor na coluna 'Class', que provavelmente representa se uma transação é fraudulenta ou não.
data['Class'].value_counts()
```

```python
# Verificando a quantidade de valores ausentes em cada coluna do conjunto de dados.
data.isnull().sum()
```

Dividindo os dados em features (x) e target (y), onde x contém todas as colunas exceto a última e y contém apenas a última coluna ('Class').

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
y
```

Dividindo os dados em conjuntos de treinamento e teste usando train_test_split, com 20% dos dados para teste e estratificação para garantir a mesma proporção de classes em ambos os conjuntos.

```python
x_train, x_test, y_train, y_test = train_test_split( x , y, test_size= 0.2, stratify= y, random_state= 42)
```

```python
# Verificando se há valores ausentes no conjunto de treinamento de (x_train).
x_train.isnull().sum()
```

```python
# Fazendo o mesmo com o conjunto alvo
y_train.isnull().sum()
```

```python
# Contando a frequência de cada valor no conjunto de treino (y_train).
y_train.value_counts()
```

```python
# Fazendo o mesmo com o conjunto alvo de testes
y_test.value_counts()
```

Criando um modelo RandomForestClassifier com pesos balanceados para as classes e o treinando com os dados de treinamento.

```python
modelo = RandomForestClassifier(class_weight='balanced', random_state=89)
modelo.fit(x_train, y_train)
```

Fazendo previsões no conjunto de teste usando o modelo treinado

```python
previsao = modelo.predict(x_test)
previsao
```

# Avaliando o desempenho do modelo.

```python
# Calculando a precisão, recall e AUC
precisao, recall, _ = precision_recall_curve(y_test, previsao)
auc = auc(recall, precisao)
auc
```

```python
 # Imprimindo o relatório de classificação, que inclui precisão, recall, F1-score e suporte para cada classe.
classificacao = classification_report(y_test, previsao)
print(classificacao)
```

 Calculando a precisão, recall e AUC e plotando a curva de precisão-recall para visualizar o desempenho do modelo.

```python
from sklearn.metrics import precision_recall_curve, auc # Make sure to import the auc function

precisao, recall, _ = precision_recall_curve(y_test, previsao)
auc_score = auc(recall, precisao) # Use the imported auc function instead of the overwritten variable

plt.plot(recall, precisao, label='AUC = {:.2f}'.format(auc_score))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curva de Precisão-Recall')
plt.legend(loc='lower left')
plt.show()
```

 Calculando a acurácia do modelo no conjunto de teste.

```python
modelo.score(x_test, y_test)
```

Criando e exibindo a matriz de confusão para visualizar o desempenho do modelo em termos de verdadeiros positivos, falsos positivos, verdadeiros negativos e falsos negativos.

```python
cm = confusion_matrix(y_test, previsao)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Previsão')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusão')
plt.show()
```
